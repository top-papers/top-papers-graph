# Task 3 — Kaggle launcher notebook for case-based blind A/B

Этот notebook больше не выполняет всю логику сам. Он подготавливает окружение и запускает **CLI runner** `scripts/task3/run_task3_case_based_blind_ab.py`.

## Что нужно

1. repo `top-papers-graph`;
2. trajectory YAML или ZIP;
3. filled case manifest JSON или ZIP от эксперта-создателя;
4. curated `processed_papers`;
5. конфиг моделей A/B.

```
from pathlib import Path
import subprocess
import sys

REPO_DIR = Path('/kaggle/working/top-papers-graph')
REPO_URL = 'https://github.com/top-papers/top-papers-graph.git'

if not REPO_DIR.exists():
    !git clone --depth 1 {REPO_URL} {REPO_DIR}

if str(REPO_DIR / 'src') not in sys.path:
    sys.path.insert(0, str(REPO_DIR / 'src'))
```

```
# Configure paths
CFG = {
    'trajectory_path': '',
    'case_manifest_path': '',
    'processed_dir': '',
    'output_dir': '/kaggle/working/task3_case_based_blind_ab',
    'query': '',
    'model_a_vlm_model_id': '',
    'model_b_vlm_model_id': '',
    'model_a_local_text_model': '',
    'model_b_local_text_model': '',
    'run_vlm': True,
    'include_multimodal': True,
    'hf_offline': False,
    'top_hypotheses': 16,
    'candidate_top_k': 24,
    'top_papers': 12,
    'edge_mode': 'auto',
    'link_prediction_backend': 'auto',
}
```

```
# Run CLI
cmd = [
    sys.executable,
    str(REPO_DIR / 'scripts' / 'task3' / 'run_task3_case_based_blind_ab.py'),
    '--trajectory', CFG['trajectory_path'],
    '--case-manifest', CFG['case_manifest_path'],
    '--processed-dir', CFG['processed_dir'],
    '--out-dir', CFG['output_dir'],
    '--query', CFG['query'],
    '--top-hypotheses', str(CFG['top_hypotheses']),
    '--candidate-top-k', str(CFG['candidate_top_k']),
    '--top-papers', str(CFG['top_papers']),
    '--edge-mode', CFG['edge_mode'],
    '--link-prediction-backend', CFG['link_prediction_backend'],
]
if CFG['model_a_vlm_model_id']:
    cmd += ['--model-a-vlm-model-id', CFG['model_a_vlm_model_id']]
if CFG['model_b_vlm_model_id']:
    cmd += ['--model-b-vlm-model-id', CFG['model_b_vlm_model_id']]
if CFG['model_a_local_text_model']:
    cmd += ['--model-a-local-text-model', CFG['model_a_local_text_model']]
if CFG['model_b_local_text_model']:
    cmd += ['--model-b-local-text-model', CFG['model_b_local_text_model']]
cmd += ['--run-vlm' if CFG['run_vlm'] else '--no-run-vlm']
cmd += ['--include-multimodal' if CFG['include_multimodal'] else '--no-include-multimodal']
cmd += ['--hf-offline' if CFG['hf_offline'] else '--no-hf-offline']

print('Running:', ' '.join(cmd))
subprocess.run(cmd, check=True, cwd=str(REPO_DIR))
```
